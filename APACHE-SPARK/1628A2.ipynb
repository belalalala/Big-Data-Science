```
shake = sc.textFile("/FileStore/tables/shakespeare_1.txt")
counts = shake.flatMap(lambda line: line.split(" "))
c2 = counts.map(lambda word: (word, 1))
c3 = c2.reduceByKey(lambda a, b: a + b)
# c3.collect()
```

```
select= c3.filter(lambda x: x[0] == 'Shakespeare' or x[0] =='When' or x[0] =='Lord' or x[0] =='Library' or x[0] =='GUTENBERG' or x[0] =='WILLIAM' or x[0] =='COLLEGE' or x[0] =='WORLD')
```

```
path = "/FileStore/tables/movies-1.csv"
mdf = spark.read \
  .format("csv") \
  .option("inferSchema", True) \
  .option("header", True) \
  .option("sep", ',') \
  .option("path", path) \
  .load()
```

```
from pyspark.ml.classification import LogisticRegression


data = mdf.select("movieId", "rating", "userId")
splits = mdf.randomSplit([0.8, 0.2])
train = splits[0].withColumnRenamed("rating", "trian")
test = splits[1].withColumnRenamed("rating", "test")
trainr = train.count()
testr = test.count()
print("Training Rows:", trainr, " Testing Rows:", testr)
```

```
lr = LogisticRegression(train, featuresCol="features", maxIter=10)
```
